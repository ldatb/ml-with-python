## About the dataset

DBSCAN is specially very good for tasks like class identification on a spatial context. The wonderful attribute of DBSCAN algorithm is that it can find out any arbitrary shape cluster without getting affected by noise. For example, this following example cluster the location of weather stations in Canada. DBSCAN can be used here, for instance, to find the group of stations which show the same weather condition. As you can see, it not only finds different arbitrary shaped clusters, can find the denser part of data-centered samples by ignoring less-dense areas or noises.

### Loading the dataset

```python
data_raw = pd.read_csv('data/weather-stations20140101-20141231.csv')
data_raw.head()
```

### Cleaning the dataset
Lets remove rows that dont have any value in the Tm field.

```python
data_raw = data_raw[pd.notnull(data_raw['Tm'])]
data_raw = data_raw.reset_index(drop = True)

data_raw.head()
```

## Clustering
Clustering of stations based on their location i.e. Lat & Lon
<br>
DBSCAN form sklearn library can runs DBSCAN clustering from vector array or distance matrix. In our case, we pass it the Numpy array Clus_dataSet to find core samples of high density and expands clusters from them.

### Requirements

```python
from sklearn.cluster import DBSCAN
import sklearn.utils
from sklearn.preprocessing import StandardScaler

sklearn.utils.check_random_state(1000)
```

### Preparing the data
We'll separate the data based on it's location

```python
cluster_data = data_raw[['Lat', 'Long']]
cluster_data = np.nan_to_num(cluster_data)
cluster_data = StandardScaler().fit_transform(cluster_data)
```

### Compute DBSCAN

```python
dbscan = DBSCAN(eps = 0.15, min_samples = 10).fit(cluster_data)
core_samples_mask = np.zeros_like(dbscan.labels_, dtype = bool)
core_samples_mask[dbscan.core_sample_indices_] = True

labels = dbscan.labels_
data_raw['Clus_Dbscan'] = labels

realClusterNumber = len(set(labels)) - (1 if -1 in labels else 0)
clusterNumber = len(set(labels))
```
